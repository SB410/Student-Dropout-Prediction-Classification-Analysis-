```python
# Question # 2: Clean data and perform feature engineering and load the package.


# Drop the specified columns from the DataFrame
clean_data = final_data.drop(columns=['2012 Loan', '2013 Loan', '2014 Loan', '2015 Loan', '2016 Loan',
                                       '2012 Scholarship', '2013 Scholarship', '2014 Scholarship', '2015 Scholarship', '2016 Scholarship',
                                       'DegreeTypeSought', 'Address1', 'Complete1', 'City', 'State'])


# Select the specified columns from the DataFrame
clean_final = clean_data[['StudentID', 'Cohort', 'CohortTerm', 'Major1', 'AcademicYear', 'TermGPA', 'CumGPA', 'Term']]

clean_final
```

```python
# Perform feature engineering on variable TermGPA by taking average and filling it with average value.

# Remove Duplicate StudentID

clean_final = clean_final.drop_duplicates(subset='StudentID', keep='last')

# Calculate the mean of 'TermGPA' excluding 0.00 values
avg = clean_final.loc[clean_final['TermGPA'] != 0.00, 'TermGPA'].mean()

# Replace 0.00 values in 'TermGPA' with the calculated average
clean_final['TermGPA'] = clean_final['TermGPA'].apply(lambda x: avg if x == 0.00 else x)

#clean_final['TermGPA']

# Perform feature engineering on variable CumGPA.

# Calculate the mean of 'CumGPA' excluding 0.00 values
avg = clean_final.loc[clean_final['CumGPA'] != 0.00, 'CumGPA'].mean()

# Replace 0.00 values in 'CumGPA' with the calculated average
clean_final['CumGPA'] = clean_final['CumGPA'].apply(lambda x: avg if x == 0.00 else x)

#clean_final['CumGPA']

# Perform feature engineering on variable Major1

# Calculate the mean of 'Major1' excluding 0.0000 values
avg = clean_final.loc[clean_final['Major1'] != 0.0000, 'Major1'].mean()

# Replace 0.0000 values in 'Major1' with the calculated average
clean_final['Major1'] = clean_final['Major1'].apply(lambda x: avg if x == 0.0000 else x)

#clean_final['Major1']

clean_final
```

```python
# Qestion # 5

# Fit various models to the training dataset by going through the usual process of further splitting this dataset into train and test and measuring performance.


# Importing train_test_split function from scikit-learn
from sklearn.model_selection import train_test_split, KFold, cross_val_score, RepeatedKFold, cross_validate

# Split the dataset into 75% training and 25% test and use 10 fold cross validation for the models below

# Separate predictors (independent variables) and target (dependent variable)
X = df.drop(['Dropout', 'AcademicYear', 'Cohort', 'StudentID'],axis=1)
X.info()

# Target (dependent variable) for binary classification
y = df['Dropout'].astype('category')

# Split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print(X_train)

```

```python
# Using Support Vector Machine to predict Student Dropout

from sklearn.svm import SVC
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

# instantiate the model
svm = SVC(kernel='rbf', C=1)

# fit the model
model = svm.fit(X_train, y_train)

# Test Accuracy
y_pred_test = model.predict(X_test)
print('Test accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))

print(confusion_matrix(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test, digits=3))
```

```python
# Plot SVM model using training dataset

from sklearn.model_selection import cross_val_score
from sklearn import metrics
import matplotlib.pyplot as plt

# Create a new SVM model
svm = SVC(kernel='rbf', C=1)

# Fit the model with training data
svm.fit(X_train, y_train)

# Calculate cross-validation score
scores = cross_val_score(svm, X_train, y_train, cv=10, scoring='accuracy')

# Print the average cross-validation score
print("Cross-validation score:", scores.mean())

# Predict labels for test data
y_pred_test = svm.predict(X_test)

# Calculate accuracy score for test data
accuracy = metrics.accuracy_score(y_test, y_pred_test)

# Print the accuracy score
print("Accuracy score:", accuracy)

# Plot the confusion matrix
plt.figure(figsize=(6, 4))
cm = metrics.confusion_matrix(y_test, y_pred_test)
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix Student DropOut')
plt.colorbar()
plt.xlabel('Student')
plt.ylabel('DropOut')
plt.show()
```

```python
# Try decision tree model on to predict whether an undergraduate will dropout or not.

# Decision Tree

# Import necessary libraries


# Create a decision tree classifier
clf = DecisionTreeClassifier(criterion='entropy', random_state=100, max_depth=3, min_samples_leaf=5)

# Train the classifier
clf = clf.fit(X_train, y_train)

# Predict the labels for the test data
y_pred = clf.predict(X_test)

# Evaluate the model accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Print the confusion matrix
print(confusion_matrix(y_test, y_pred))

# Print the classification report
print(classification_report(y_test, y_pred))

# Visualize the decision tree
plt.figure(figsize=(10, 7))
plot_tree(clf, feature_names=X.columns, class_names=['Not Dropout', 'Dropout'], filled=True)
plt.show()
```
